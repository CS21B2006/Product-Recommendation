```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
path = '/content/Drive/MyDrive/online retail.xlsx'
df = pd.read_excel(path)
df
```

```python
df.dropna(inplace=True)
```

```python
df1=df.drop(columns=['InvoiceNo','Description','Quantity','UnitPrice','Country'])
```

```python
df1
```

```python
df1["InvoiceDate"] = df1["InvoiceDate"].dt.strftime('%Y')
```

```python
df1
```

```python
features = ['InvoiceDate', 'CustomerID']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df1[features] = scaler.fit_transform(df1[features])
```

```python
from sklearn.cluster import KMeans
k = 5  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(df1[features])
```

```python
# Add cluster labels to the dataset
df1['cluster'] = kmeans.labels_
```

```python
df1
```

```python
stock = df1.loc[df1['cluster'] == 4, 'StockCode'].values
print(stock)
```

```python
import random
def predict_cluster(features):
    scaled_features = scaler.transform([features])
    cluster = kmeans.predict(scaled_features)[0]
    if cluster in df1['cluster'].values:
        stock_code = df1.loc[df1['cluster'] == cluster, 'StockCode'].values    
    li=[]    
    rand_idx = random.randint(0, len(stock_code)-1)
    li.append(stock_code[rand_idx])
    print("The Recommended Product :",*li)
```

```python
customer_id = '12680.0'
date = pd.to_datetime('2011-12-01 08:26:00').date()
dt=date.year
fea=[]
fea.append(customer_id)
fea.append(dt)
predict_cluster(fea)
```
